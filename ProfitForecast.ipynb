```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
# Ler a aba "Receita"
df_receita = pd.read_excel('fluxo_caixa_cbs.xlsx', sheet_name='Receita')

# Ler a aba "Despesa"
df_despesa = pd.read_excel('fluxo_caixa_cbs.xlsx', sheet_name='Despesa')
```

```python
df_despesa
```

```python
df_receita
```

```python
# Agrupando por 'Tipo' e somando os 'Valor'
result_receita = df_receita.groupby('Tipo')['Valor'].sum().reset_index()
result_despesa = df_despesa.groupby('Tipo')['Valor'].sum().reset_index()
```

```python
result_despesa
result_receita
```

```python
result_despesa
```

```python
print(result_despesa)
print(result_receita)
```

```python
despesa_agrupada = result_despesa.groupby('Tipo')['Valor'].sum().reset_index()

# Renomear a coluna para maior clareza
despesa_agrupada.columns = ['Descrição', 'Total']

# Exibir a matriz agrupada
print("Matriz de Despesa Agrupada:")
print(despesa_agrupada)
```

```python
receita_agrupada = result_receita.groupby('Tipo')['Valor'].sum().reset_index()

# Renomear a coluna para maior clareza
receita_agrupada.columns = ['Tipo', 'ValorTotal']

# Exibir a matriz agrupada
print("Matriz de Receita Agrupada:")
print(receita_agrupada)
```

```python
# Criar o gráfico de barras
fig = go.Figure()
fig.add_trace(go.Bar(
    x=despesa_agrupada['Descrição'],
    y=despesa_agrupada['Total'],
    marker_color='#ff7b00'  # Cor laranja da barra
))

# Personalizar o layout do gráfico
fig.update_layout(
    title='Soma dos Valores por Tipo de Despesa',
    xaxis_title='Tipo',
    yaxis_title='Valor_Total',
    font_family='Arial',
    font_color='#333333',
    title_font_size=20,
    xaxis_tickfont_size=14,
    yaxis_tickfont_size=14,
    bargap=0.1,
    plot_bgcolor='#f5f5f5',
    paper_bgcolor='#f5f5f5'
)

# Exibir o gráfico
fig.show()
```

```python
# Criar o gráfico de barras
fig = go.Figure()
fig.add_trace(go.Bar(
    x=receita_agrupada['Tipo'],
    y=receita_agrupada['ValorTotal'],
    marker_color='#ff7b00'  # Cor laranja da barra
))

# Personalizar o layout do gráfico
fig.update_layout(
    title='Soma dos Valores por Tipo de Despesa',
    xaxis_title='Tipo',
    yaxis_title='Valor_Total',
    font_family='Arial',
    font_color='#333333',
    title_font_size=20,
    xaxis_tickfont_size=14,
    yaxis_tickfont_size=14,
    bargap=0.1,
    plot_bgcolor='#f5f5f5',
    paper_bgcolor='#f5f5f5'
)

# Exibir o gráfico
fig.show()
```

```python
# Criar o gráfico de barras
fig = go.Figure()

# Adicionar as barras de despesas
fig.add_trace(go.Bar(
    x=despesa_agrupada['Descrição'],  # ou 'Tipo' se for o caso
    y=despesa_agrupada['Total'],
    name='Despesas',
    marker_color='#ff7b00'  # Cor laranja da barra
))

# Adicionar as barras de receitas
fig.add_trace(go.Bar(
    x=receita_agrupada['Tipo'],  # ou 'Descrição' se for o caso
    y=receita_agrupada['ValorTotal'],
    name='Receitas',
    marker_color='#007bff'  # Cor azul da barra
))

# Personalizar o layout do gráfico
fig.update_layout(
    title='Comparação entre Receitas e Despesas',
    xaxis_title='Tipo',
    yaxis_title='Valor',
    barmode='group',  # Agrupar as barras
    font_family='Arial',
    font_color='#333333',
    title_font_size=20,
    xaxis_tickfont_size=14,
    yaxis_tickfont_size=14,
    bargap=0.1,
    plot_bgcolor='#f5f5f5',
    paper_bgcolor='#f5f5f5'
)

# Exibir o gráfico
fig.show()
```
